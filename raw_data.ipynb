#### Import necessary libraries

```python
import pandas as pd
```

#### Read the dataframe

```python
df = pd.read_csv("D:/Bachelors/Semester2/VDSS/project_folder/Student_Mental_health.csv")
df
```

#### Clean up the data

```python
# Dropping not necessary columns
df=df.drop(columns='Timestamp')
df
```

```python
#replacing values in columns
mapping_dict = {'year 1': 1,'Year 1': 1,
                'year 2': 2,'Year 2': 2,
                'year 3': 3,'Year 3': 3,
                'year 4': 4,'Year 4': 4}
df = df.replace(mapping_dict)
df
```

```python
mapping_dict1 = {'Female': 'F','Male': 'M'}
df = df.replace(mapping_dict1)
df
```

```python
#Droping rows of NAN
nan_df = df[df.isna().any(axis=1)]
display(nan_df.head())

df = df.dropna(how='any')
df
```

```python
#Converting Age into int;removing the decimals
df = df.astype({'Age':'int'})
df
```

```python
#Changing the names of some columns
df.rename(columns = {'Choose your gender':'Gender',
                     'What is your course?':'Major',
                     'What is your CGPA?':'CGPA',
                     'Your current year of Study':'Year',
                     'Do you have Depression?':'Depression',
                     'Do you have Anxiety?': 'Anxiety',
                     'Do you have Panic attack?':'Panic attacks',
                     'Did you seek any specialist for a treatment?':'Therapy',
                    'Marital status':'Single'}, inplace = True)
df
```
