```python
df = pd.read_csv('D:\Bachelors\Semester2\VDSS\project_folder\VDSS_Visualisierungsprojekt\clean_data\\2019.csv')
df
```

```python
df.rename(columns = {'Overall rank':'Happiness Rank','Score':'Happiness Score','GDP per capita':'Economy (GDP per Capita)','Healthy life expectancy': 'Health (Life Expectancy)','Perceptions of corruption':'Trust (Government Corruption)','Country or region':'Country'}, inplace = True)
df
```

```python
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df

#swap points and rebounds columns
df = swap_columns(df, 'Country','Happiness Rank')
df = swap_columns(df, 'Trust (Government Corruption)','Generosity')
df
```

```python
df=df.drop(columns='Generosity')
df
```
