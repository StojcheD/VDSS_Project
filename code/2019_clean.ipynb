```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
df = pd.read_csv(path + '\..\\raw_data\\World Happiness\\2019.csv')
df
```

```python
df.rename(columns = {'Overall rank':'Happiness Rank','Score':'Happiness Score','GDP per capita':'Economy (GDP per Capita)','Healthy life expectancy': 'Health (Life Expectancy)','Perceptions of corruption':'Trust (Government Corruption)','Country or region':'Country'}, inplace = True)
df
```

```python
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df

#swap points and rebounds columns
df = swap_columns(df, 'Country','Happiness Rank')
df = swap_columns(df, 'Trust (Government Corruption)','Generosity')
df
```

```python
df=df.drop(columns='Generosity')
df
```

```python
nan_df = df[df.isna().any(axis=1)]
display(nan_df.head())

df = df.dropna(how='any')
df
```

```python
region_df = pd.read_csv(path + '\..\\raw_data\\World Happiness\\2015.csv')

# Merge the DataFrames based on "Country" column
df = pd.merge(df, region_df[['Country', 'Region']], on='Country', how='left')

# Reorder the columns with "Region" in the third column position
df = df[df.columns[:2].tolist() + ['Region'] + df.columns[2:-1].tolist()]

df
```

```python
# Filter the DataFrame to get rows with missing values in the "Region" column
missing_region_countries = df[df['Region'].isna()]['Country'].tolist()

# Print the countries with missing region
for country in missing_region_countries:
    print(country)
```

```python
# Create a dictionary mapping countries to their respective regions
region_mapping = {
    'Trinidad & Tobago': 'Latin America and Caribbean',
    'Northern Cyprus': 'Middle East and Northern Africa',
    'North Macedonia': 'Central and Eastern Europe',
    'Somalia': 'Sub-Saharan Africa',
    'Namibia': 'Sub-Saharan Africa',
    'Gambia': 'Sub-Saharan Africa',
    'South Sudan': 'Sub-Saharan Africa'
}

# Add the missing "Region" values for the specified countries
for country, region in region_mapping.items():
    df.loc[df['Country'] == country, 'Region'] = region
```

```python
# Move "Happiness Rank" to the first column
df = df[['Happiness Rank'] + df.columns.drop('Happiness Rank').tolist()]
```

```python
df.to_csv(path + "/../clean_data/2019.csv", index=False)
```
