```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
df = pd.read_csv(path + '\..\\raw_data\\World Happiness\\2020.csv')
df
```

```python
df = df.drop(df.loc[:, 'Ladder score in Dystopia':'Dystopia + residual'].columns, axis=1)
```

```python
df=df.drop(columns=['Standard error of ladder score','upperwhisker','lowerwhisker','Generosity',])
df
```

```python
df.rename(columns = {'Country name':'Country','Regional indicator':'Region','Ladder score':'Happiness Score','Logged GDP per capita':'Economy (GDP per Capita)','Healthy life expectancy':'Health (Life Expectancy)','Perceptions of corruption':'Trust (Government Corruption)'}, inplace = True)
df
```

```python
nan_df = df[df.isna().any(axis=1)]
display(nan_df.head())

df = df.dropna(how='any')
df
```

```python
# Add a new column "Ranking" based on "Happiness Score"
df['Happiness Rank'] = df['Happiness Score'].rank(ascending=False).astype(int)
# Reorder the columns with "Ranking" as the first column
df = df[['Happiness Rank'] + df.columns[:-1].tolist()]
df
```

```python
# Loop through each column in the DataFrame
for col in df.columns:
    # Check if the column is of float type
    if df[col].dtype == 'float64':
        # Round the values in the column to three decimal places
        df[col] = df[col].round(3)
        
df
```

```python
df.to_csv(path + "/../clean_data/2020.csv", index=False)
```
