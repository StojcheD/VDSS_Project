```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = os.getcwd()
raw = pd.read_csv(path + "/../raw_data/global_mental_health.csv", low_memory=False)

```

```python
old_df = raw.loc[:,["COUNTRYNEW", "Global11Regions", "MH1", "MH7A", "Age", "age_var3", "Gender", "Education", "wbi"]]
old_df
```

```python
df = old_df
df.rename(columns = {"COUNTRYNEW":"Country",
                     "Global11Regions":"Region",
                     "MH1":"Importance",
                     "MH7A":"Self_experience",
                     "age_var3":"Age_range",
                     "wbi":"Country_wealth"
                     }, inplace=True)

df
```

```python
nan_df = df[df.isna().any(axis=1)]
nan_df # 0 NA values
df.info()
```

```python
df["Region"] = df["Region"].replace({1: "Western Europe",
                                             2: "Eastern Europe",
                                             3: "Post-Soviet Eurasia",
                                             4: "Australia",
                                             5: "East Asia",
                                             6: "South-East Asia",
                                             7: "South Asia",
                                             8: "South America",
                                             9: "North America",
                                             10: "North Africa",
                                             11: "Sub-Saharan Africa"})
df
```

```python
df["Self_experience"] = df["Self_experience"].replace({"1": "Yes", "2": "No", "99": "DK/Refused"})
df
```

```python
df["Importance"] = df["Importance"].replace({1: "More important", 2: "As important", 3: "Less important", 99: "DK/Refused"})
df
```

```python
df["Age_range"] = df["Age_range"].replace({1: "15-24",
                                           2: "25-34",
                                           3: "35-49",
                                           4: "50+",
                                           99: "DK/Refused"})
df
```

```python
df["Gender"] = df["Gender"].replace({1: "M",2: "F"})
df
```

```python
df["Education"] = df["Education"].replace({1: "Elementary or less",
                                           2: "Secondary",
                                           3: "Tertiary",
                                           })
df
```

```python
df["Country_wealth"] = df["Country_wealth"].replace({1: "Low income",
                                           2: "Lower-middle income",
                                           3: "Upper-middle income",
                                           4: "High income",
                                           })
df
```

```python
print(df.dtypes)
```

```python
# Iterate through each column in the DataFrame
for column in df.columns:
    # Check if the column is of string data type
    if df[column].dtype == 'object':
        # Remove rows containing "DK/Refused" in the column
        df = df[~(df[column] == 'DK/Refused')]
```

```python
df.to_csv(path + "/../clean_data/cleaned_data_global_mental_health.csv", index = False)
```
