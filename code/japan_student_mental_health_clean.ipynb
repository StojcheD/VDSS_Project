```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
df = pd.read_csv(path + '\..\\raw_data\\data.csv')
df
```

```python
df = df.drop(df.loc[:, 'ToSC':'AMiscell'].columns, axis=1)
df = df.drop(df.loc[:, 'Partner':'Internet_bi'].columns, axis=1)
```

```python
df=df.drop(columns=['Stay_Cate','Japanese','Japanese_cate','Intimate','Religion','Stay','Age_cate','English','English_cate'])
df
```

```python
nan_df = df[df.isna().any(axis=1)]
display(nan_df.head())

df = df.dropna(how='any')
df
```

```python
df = df.astype({'Age':'int'})
df = df.astype({'ToDep':'int'})
df = df.astype({'ToAS':'int'})
df
```

```python
df.rename(columns = {'Dep':'Depression','Suicide':'Suicidal Ideation','ToDep':'Depression score','ToAS':'Acculturative Stress'}, inplace = True)
df
```

```python
df.to_csv(path + "/../clean_data/japan_student_mental_health.csv", index=False)
```
