#### Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

#### Read the dataframe

```python
path = os.getcwd()
df = pd.read_csv(path + '\..\\raw_data\\malaysia_student_mental_health.csv')
df
```

#### Clean up the data

```python
# Dropping not necessary columns
df=df.drop(columns='Timestamp')
df
```

```python
#replacing values in columns
mapping_dict = {'year 1': 1,'Year 1': 1,
                'year 2': 2,'Year 2': 2,
                'year 3': 3,'Year 3': 3,
                'year 4': 4,'Year 4': 4}
df = df.replace(mapping_dict)
df
```

```python
#Droping rows of NAN
nan_df = df[df.isna().any(axis=1)]
display(nan_df.head())

df = df.dropna(how='any')
df
```

```python
#Converting Age into int;removing the decimals
df = df.astype({'Age':'int'})
df
```

```python
#Changing the names of some columns
df.rename(columns = {'Choose your gender':'Gender','What is your course?':'Major','What is your CGPA?':'CGPA','Your current year of Study':'Year','Do you have Depression?':'Depression','Do you have Anxiety?': 'Anxiety','Do you have Panic attack?':'Panic_attacks','Did you seek any specialist for a treatment?':'Therapy','Marital status':'Single'}, inplace = True)
df
```

#### Save cleaned data

```python
df.to_csv(path + "/clean_data/malaysia_clean_mental_health.csv")
```

```python
#making a new data frame with only numerical values so its easier to analyze
mapping_dict_0 = {'Yes': 1,'No': 0}
df_analyze = df.replace(mapping_dict_0)
df_analyze
```

```python
#adding a new column for the mental health
conditions =[
    (df_analyze['Depression'] + df_analyze['Anxiety'] + df_analyze['Panic_attacks'] == 3),
    (df_analyze['Depression'] + df_analyze['Anxiety'] + df_analyze['Panic_attacks'] == 2),
    (df_analyze['Depression'] + df_analyze['Anxiety'] + df_analyze['Panic_attacks'] == 1),
    (df_analyze['Depression'] + df_analyze['Anxiety'] + df_analyze['Panic_attacks'] == 0)
]

values = ['very_bad','bad','good','very good']

df_analyze['mental_health'] = np.select(conditions,values)
df_analyze
```

#### correlation between the grade and mental health

```python
df_4['CGPA'] = df_4['CGPA'].str.replace(r'\D', '', regex=True).astype(int)
```
